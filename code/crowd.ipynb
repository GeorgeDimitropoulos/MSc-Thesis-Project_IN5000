```python
me=[]
for index, row in data.iterrows():
    
    str=row['annotation']
    egw=re.sub(r'\b\w{4,35}\b', '', str)
    newstr = ''.join((ch if ch in '0123456789' else ' ') for ch in egw)
    listOfNumbers = [int(i) for i in newstr.split()]
    if (len(listOfNumbers)==4):
        temp1=data["elements"][index]
        str1=''.join(temp1)
        listOfNumbers.insert(0, str1)
        me.append(listOfNumbers)
    else:
        kalos=[listOfNumbers[i:i + 4] for i in range(0, len(listOfNumbers), 4)]
        for i in range(len(kalos)):
            temp2=data["elements"][index][i]
            str2 = ''.join(temp2)
            kalos[i].insert(0, str2)
        me.append(kalos)
```
